```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import torch
import torch.nn as nn
from torch.autograd import Variable
import numpy as np
```

```python
training_set = pd.read_csv('./monthly-lake-erie-levels-1921-19.csv')
training_set = training_set.iloc[:,1:2].values


plt.plot(training_set)
plt.title('Monthly lake erie levels') #伊利湖水平線
plt.show()
```

```python
from sklearn.preprocessing import MinMaxScaler #透過MinMaxScaler轉換到0~1
sc = MinMaxScaler() 
training_data = sc.fit_transform(training_set)
```

```python
def sliding_windows(data, seq_length):
    x = []
    y = []

    for i in range(len(data)-seq_length-1): #為了能夠依據seq_length做截斷的做動作
        _x = data[i:(i+seq_length)] #訓練集，依據seq_length取一段
        _y = data[i+seq_length] #預測值每，依據seq_length取一段
        x.append(_x)
        y.append(_y)

    return np.array(x),np.array(y)
```

```python
seq_length = 7 #長度
learning_rate = 0.01 #學習率
input_size = 1 #輸入層
hidden_size = 5 #隱藏層
num_classes = 1 #分類數量
num_layers = 1 #LSTM層數
num_epochs = 400 #迭代
```

```python
x, y = sliding_windows(training_data, seq_length) #每seq_length一段

train_size = int(len(y) *0.7) #設定訓練集大小，train_size大小會影響到圖的密集程度，數值乘越大越鬆
test_size = len(y) - train_size #設定測試集大小
trainX = torch.Tensor(np.array(x[0:train_size]))
trainX = Variable(trainX) #自動微分
testX = torch.Tensor(np.array(x[train_size:len(x)]))
testX = Variable(testX) #自動微分
trainY = torch.Tensor(np.array(y[0:train_size]))
trainY = Variable(trainY) #自動微分
testY = torch.Tensor(np.array(y[train_size:len(y)]))
testY = Variable(testY) #自動微分
```

```python
class LSTM(nn.Module):

    def __init__(self, num_classes, input_size, hidden_size, num_layers):
        super(LSTM, self).__init__()
        self.num_classes = num_classes
        self.num_layers = num_layers
        self.input_size = input_size
        self.hidden_size = hidden_size
        self.seq_length = seq_length
        self.lstm = nn.LSTM(input_size=input_size, hidden_size=hidden_size,
                            num_layers=num_layers, batch_first=True)
        self.fc = nn.Linear(hidden_size, num_classes) #建立限性轉換

    def forward(self, x):
        h_0 = Variable(torch.zeros(self.num_layers, x.size(0), self.hidden_size))
        c_0 = Variable(torch.zeros(self.num_layers, x.size(0), self.hidden_size))
        # Propagate input through LSTM
        ula, (h_out, _) = self.lstm(x, (h_0, c_0))
        h_out = h_out.view(-1, self.hidden_size)
        out = self.fc(h_out)
        return out
```

```python
lstm = LSTM(num_classes, input_size, hidden_size, num_layers)

criterion = torch.nn.MSELoss()    # mean-squared error for regression
optimizer = torch.optim.Adam(lstm.parameters(), lr=learning_rate) #使用Adam優化器

# 訓練模型
for epoch in range(num_epochs):
    outputs = lstm(trainX)
    optimizer.zero_grad()
    # 回傳Loss
    loss = criterion(outputs, trainY)
    loss.backward()
    optimizer.step()
    #if epoch % 100 == 0:
    print("Epoch: %d, loss: %1.5f" % (epoch, loss.data[0]))
```

```python
lstm.eval()
test_predict = lstm(testX)

test_predict = test_predict.data.numpy()
testY = testY.data.numpy()
plt.plot(testY)
plt.plot(test_predict)
plt.show()
```
